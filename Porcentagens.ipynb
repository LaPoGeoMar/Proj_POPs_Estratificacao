```python
import pandas as pd
```

```python
dataframe = pd.read_excel("teste.xlsx", sheet_name = 1, index_col = 0)
dataframe
```

```python
stacked_data = dataframe.apply(lambda x: x*100/sum(x), axis=1)
stacked_data.plot(kind="bar", stacked=True)
```

```python
dataframe.plot.bar(stacked=True); 
```

```python
data1 = pd.read_excel("teste.xlsx", sheet_name = 2, index_col = 0)
data1
```
