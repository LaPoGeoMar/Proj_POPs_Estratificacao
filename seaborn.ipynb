```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_excel("Resumo.xlsx")
dataframe
```

```python
PCBs = dataframe["PCBs"]
DDTs = dataframe["DDTs"]
CHLs = dataframe["CHLs"]
HCB = dataframe["HCB"]
Dieldrin = dataframe["Dieldrin"]
Mirex = dataframe["Mirex"]
camada = dataframe["Camada"]
```

```python
colors = ["red", "blue"]
fig, ax = plt.subplots(figsize=(10, 10))
sns.set_style("white")
sns.set(font_scale = 1)
sns.boxplot(x=camada, y=PCBs, palette=colors, ax=ax)
plt.show()
```
