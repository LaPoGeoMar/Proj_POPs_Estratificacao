```python
import pandas as pd
df = pd.read_excel("OCs_24Out2021.xls", sheet_name = 3, index_col=0).T
df
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
values1 = df["Somatório Clordanos"]

stats.probplot(values1, dist="norm", plot=plt)
plt.title('C')
plt.ylabel('Quantis observados')
plt.xlabel('Quantis teóricos')
plt.show()
```

```python
from scipy.stats import shapiro
data = df["Somatório Clordanos"]
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f\n' % (stat,p))
if p > 0.05:
    print("Probably Gaussian")
else:
    print("Probably Not Gaussian")
```
